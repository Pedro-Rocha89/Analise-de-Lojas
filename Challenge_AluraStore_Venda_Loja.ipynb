### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

### **Relatório de Análise para Venda de Loja**



* Introdução


---



Este relatório tem como objetivo analisar o desempenho de quatro lojas distintas com base em dados de faturamento, vendas por categoria, avaliações de clientes, produtos mais e menos vendidos e custo de frete. A análise visa fornecer uma recomendação embasada ao Senhor João sobre qual loja ele deveria considerar vender, buscando otimizar seu portfólio de negócios.


* Análise dos Dados


---



Para embasar a decisão, foram extraídas e visualizadas as seguintes métricas para cada loja:



1.   Faturamento Total



O faturamento total é um indicador crucial de desempenho.

Loja 1: R$ 1.534.509,12

Loja 2: R$ 1.488.459,06

Loja 3: R$ 1.464.025,03

Loja 4: R$ 1.384.497,58


Conforme o "*Gráfico de Análise de Faturamento*", a Loja 1 apresenta o maior faturamento, seguida pela Loja 2. A Loja 4 possui o menor faturamento entre as quatro. Isso já nos dá uma primeira indicação de qual loja está gerando mais receita bruta.



2.   Vendas por Categoria de Produto


A diversidade e o volume de vendas por categoria indicam a força do portfólio de cada loja.

Loja 1: Lidera em vendas de Móveis (465 unidades) e Eletrônicos (448 unidades).

Loja 2: Similar à Loja 1, com Móveis (442 unidades) e Eletrônicos (422 unidades) como as categorias mais vendidas.

Loja 3: Apresenta o maior volume em Móveis (499 unidades) e Eletrônicos (451 unidades).

Loja 4: Também com forte desempenho em Móveis (480 unidades) e Eletrônicos (451 unidades).

Os gráficos de "*Análise Vendas x Categoria*" para cada loja revelam uma consistência: Móveis e Eletrônicos são as categorias dominantes em todas as lojas, indicando uma forte demanda nesses setores. A Loja 3 se destaca ligeiramente no volume de vendas de móveis.

3.   Média de Avaliação dos Clientes

A avaliação média reflete a satisfação do cliente, o que é vital para a reputação e vendas futuras.

Loja 1: 3.98 pontos

Loja 2: 4.04 pontos

Loja 3: 4.05 pontos

Loja 4: 4.00 pontos

O "*Gráfico de Análise da Avaliação de Clientes*" mostra que as Lojas 3 e 2 possuem as maiores médias de avaliação, indicando uma ligeira vantagem na satisfação do cliente. A Loja 1 tem a avaliação mais baixa, embora a diferença entre todas as lojas seja pequena, o que sugere um nível de serviço e produto relativamente consistente.

4.  Produtos Mais e Menos Vendidos

Analisar os produtos de maior e menor giro ajuda a entender o estoque e a demanda.

Loja 1:

Menos vendidos: Celular ABXY, Headset e Panela de pressão.

Loja 3:

Mais vendidos: Kit banquetas, Mesa de jantar e Cama king.

Menos vendidos: Blocos de montar, Jogo de copos e Micro-ondas.

Loja 4:

Mais vendidos: Cama box, Faqueiro e Cama king.

Menos vendidos: Guitarra, Guarda roupas e Violão.


Observa-se que, embora "Eletrônicos" e "Móveis" sejam categorias fortes em geral (conforme visto na análise de categorias), a performance individual dos produtos dentro dessas categorias pode variar bastante entre as lojas. Por exemplo, o Micro-ondas é um dos mais vendidos na Loja 1 e 2, mas um dos menos vendidos na Loja 3. Essa granularidade é importante para entender as especificidades do perfil de consumo de cada loja. A Loja 2 apresenta uma peculiaridade com o livro "Iniciando em Programação" entre seus produtos mais vendidos, o que pode indicar um nicho específico ou uma demanda local interessante.



5.  Frete Médio por Loja

O custo do frete pode impactar a competitividade e a percepção do cliente.

Loja 1: R$ 34,69

Loja 2: R$ 33,62

Loja 3: R$ 33,07

Loja 4: R$ 31,28

O "*Gráfico de Análise do Custo do Frete*" demonstra que a Loja 4 possui o frete médio mais baixo, o que pode ser um atrativo para clientes e potencialmente impulsionar vendas, especialmente para produtos com frete sensível. A Loja 1, por outro lado, tem o frete médio mais alto.

* Conclusão e Recomendação

---


Considerando a análise dos dados, a recomendação para o Senhor João seria manter as Lojas 1, 2 e 3, e considerar a venda da Loja 4.


* Justificativa:


Loja 1 (Manter): Apesar de ter o frete médio mais alto e a menor média de avaliação, a Loja 1 é a que apresenta o maior faturamento total. Isso indica que, mesmo com esses pontos fracos, ela consegue gerar a maior receita, o que é fundamental para a saúde financeira do negócio. Uma análise mais aprofundada poderia investigar se o frete mais alto é compensado por outros fatores (como a localização que permite um maior volume de vendas ou produtos de maior margem).

Loja 2 (Manter): A Loja 2 possui o segundo maior faturamento e a segunda maior média de avaliação de clientes (4.04 pontos). Ela demonstra um equilíbrio saudável entre geração de receita e satisfação do cliente. A presença de produtos como "Iniciando em Programação" entre os mais vendidos também pode indicar um nicho de mercado promissor que vale a pena explorar e desenvolver.

Loja 3 (Manter): A Loja 3 se destaca por ter a maior média de avaliação de clientes (4.05 pontos) e um volume muito bom de vendas de móveis e eletrônicos. Embora seu faturamento seja um pouco menor que o da Loja 2, a alta satisfação do cliente é um ativo valioso que pode garantir a fidelidade e o crescimento a longo prazo, além de um custo de frete competitivo.

Loja 4 (Vender): A Loja 4 possui o menor faturamento total entre todas as lojas e uma média de avaliação intermediária. Apesar de ter o frete médio mais baixo, isso não se traduziu em um faturamento superior. Se o objetivo é otimizar o portfólio e focar nos negócios mais lucrativos ou com maior potencial, a Loja 4 seria a candidata mais lógica para a venda, pois seu desempenho financeiro atual é o mais fraco.

* Observações Finais:

A decisão de vender uma loja não deve se basear apenas nos dados financeiros atuais. Fatores como o potencial de crescimento futuro em determinadas regiões, o custo-benefício de melhorias na satisfação do cliente, ou a necessidade de capital para investimentos em outras lojas, também poderiam influenciar a decisão final do Senhor João. No entanto, com base exclusivamente nos dados de desempenho analisados, a Loja 4, com o menor faturamento e um desempenho que não se destaca em outras métricas relevantes, é a candidata mais indicada para a venda.
